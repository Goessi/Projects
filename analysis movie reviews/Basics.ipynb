```python
import pandas as pd
movies = pd.read_csv('fandango_score_comparison.csv')
movies
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(movies['Metacritic_norm_round'])
plt.show()
plt.hist(movies['Fandango_Stars'])
plt.show()
```

```python
#Fandango_Stars has no stars less than 3 while Metacritic_norm_round 
#distributes normly.
```

```python
import numpy
a = movies['Fandango_Stars'].mean()
b = movies['Metacritic_norm_round'].mean()
c = movies['Fandango_Stars'].median()
d = movies['Metacritic_norm_round'].median()
e = movies['Fandango_Stars'].std()
f = movies['Metacritic_norm_round'].std()
```

```python
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
```

```python
#we could see from above, the Metacritic_norm_round mean is larger than its median
#the Fandango_Stars mean is less than its median
#that's because Metacritic_norm_round has less big numbers
# and also, there are more larger scores in Fandango_Stars
```

```python
from scipy.stats import pearsonr
r,p = pearsonr(movies['Fandango_Stars'],movies['Metacritic_norm_round'])
from scipy.stats import linregress
rss = linregress(movies['Metacritic_norm_round'],movies['Fandango_Stars'])
```

```python
print(r)
print(rss)
```

```python
pred_3 = 3*rss[0] + rss[1]
pred_3
```

```python
pred_1 = 1*rss[0] + rss[1]
pred_5 = 5*rss[0] + rss[1]
x = [1,5]
y = [pred_1,pred_5]
plt.xlim(1,5)
plt.scatter(x,y)
plt.show()
plt.plot(x,y)
plt.show()
```
