```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv',encoding = 'ISO-8859-1')
a = pd.notnull(star_wars['RespondentID'])
a.head()
star_wars = star_wars[a]
star_wars.head()
```

```python
y_n = {'Yes':True,'No':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(y_n)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(y_n)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']
```

```python
import numpy as np
see1 = {np.nan:False, 'Star Wars: Episode I  The Phantom Menace':True}
see2 = {np.nan:False, 'Star Wars: Episode II  Attack of the Clones':True}
see3 = {np.nan:False, 'Star Wars: Episode III  Revenge of the Sith':True}
see4 = {np.nan:False, 'Star Wars: Episode IV  A New Hope':True}
see5 = {np.nan:False, 'Star Wars: Episode V The Empire Strikes Back':True}
see6 = {np.nan:False, 'Star Wars: Episode VI Return of the Jedi':True}

star_wars[star_wars.columns[3]] = star_wars[star_wars.columns[3]].map(see1)
star_wars[star_wars.columns[4]] = star_wars[star_wars.columns[4]].map(see2)
star_wars[star_wars.columns[5]] = star_wars[star_wars.columns[5]].map(see3)
star_wars[star_wars.columns[6]] = star_wars[star_wars.columns[6]].map(see4)
star_wars[star_wars.columns[7]] = star_wars[star_wars.columns[7]].map(see5)
star_wars[star_wars.columns[8]] = star_wars[star_wars.columns[8]].map(see6)

```

```python
star_wars = star_wars.rename(columns = {'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
                                        'Unnamed: 4':'seen_2',
                                        'Unnamed: 5':'seen_3',
                                        'Unnamed: 6':'seen_4',
                                        'Unnamed: 7':'seen_5',
                                        'Unnamed: 8':'seen_6'})
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns = {'Please rank the Star Wars films in order of preference with 1 being your least favourite film':'ranking_1',
                                        'Unnamed: 10':'ranking_2',
                                        'Unnamed: 11':'ranking_3',
                                        'Unnamed: 12':'ranking_4',
                                        'Unnamed: 13':'ranking_5',
                                        'Unnamed: 14':'ranking_6'})
star_wars.head()
```
