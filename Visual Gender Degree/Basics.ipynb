```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']


fig = plt.figure(figsize=(16,20))

# first column
for num in range(0,18,3):
    index = int(num/3)
    ax = fig.add_subplot(6,3,num+1)
    ax.plot(women_degrees['Year'],women_degrees[stem_cats[index]],c = cb_dark_blue,linewidth = 3)
    ax.plot(women_degrees['Year'],100 - women_degrees[stem_cats[index]],c = cb_orange,linewidth = 3)
    ax.tick_params(labelbottom = 'off',top = 'off',bottom = 'off',left = 'off',right = 'off')
    if index == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom = 'on')
    for i,v in ax.spines.items():
        v.set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[index])
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/256,171/256,171/256),alpha=0.3)
#second column
lib_arts_cats = ['Foreign Languages','English','Communications and Journalism',
                 'Art and Performance','Social Sciences and History']
for num in range(1,16,3):
    index = int((num - 1)/3)
    ax = fig.add_subplot(6,3,num+1)
    ax.plot(women_degrees['Year'],women_degrees[lib_arts_cats[index]],c = cb_dark_blue,linewidth = 3)
    ax.plot(women_degrees['Year'],100 - women_degrees[lib_arts_cats[index]],c = cb_orange,linewidth = 3)
    for i,v in ax.spines.items():
        v.set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[index])
    ax.tick_params(labelbottom = 'off',top = 'off',bottom = 'off',left = 'off',right = 'off')
    if index == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif index == 4:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom = 'on')
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/256,171/256,171/256),alpha=0.3)
#third column
other_cats = ['Health Professions','Public Administration','Education','Agriculture','Business','Architecture']
for num in range(2,18,3):
    index = int((num-2)/3)
    ax = fig.add_subplot(6,3,num+1)
    ax.plot(women_degrees['Year'],women_degrees[other_cats[index]],c = cb_dark_blue,linewidth = 3)
    ax.plot(women_degrees['Year'],100 - women_degrees[other_cats[index]],c = cb_orange,linewidth = 3)
    ax.set_xlim = (1968,2011)
    ax.set_ylim = (0,100)
    ax.set_title(other_cats[index])
    for k,v in ax.spines.items():
        v.set_visible(False)
    ax.tick_params(labelbottom = 'off',top = 'off',bottom = 'off',left = 'off',right = 'off')
    if index == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom = 'on')
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/256,171/256,171/256),alpha=0.3)
    plt.savefig('gender_degrees.png')
```
