```python
import csv
f=open('AviationData.txt',encoding='utf-8')
aviation_data=f.readlines()
aviation_data
```

```python
aviation_list = []
for line in aviation_data:
    line = line.split(' | ')
    aviation_list.append(line)
aviation_list
```

```python
lax_code = []
for m in range(len(aviation_list)):
    line = aviation_list[m]
    for n in range(len(line)):
        if line[n] == 'LAX94LA336':
            lax_code.append(line)
lax_code
```

The way above loop through every row, every column

```python
# linear time          
lax_code = []
for line in aviation_list:
    if 'LAX94LA336' in line:
        lax_code.append(line)
lax_code
```

```python
# log(n) algorithm
import bisect
sorted_aviation_list = sorted(aviation_list, key = lambda row: row[2])
sorted_accident_numbers = [row[2] for row in sorted_aviation_list]
lax_index = bisect.bisect_left(sorted_accident_numbers, "LAX94LA336")
print(sorted_aviation_list[lax_index])
```

```python
aviation_dict_list = []
head = aviation_data[0]
head
```

```python
aviation_data
```

```python
head = head.split(' | ')
aviation_data = aviation_data[1:]
for line in aviation_data:
    d = {}
    line = line.split(' | ')
    for i in range(len(line) - 1):
        key = head[i]
        value = line[i]
        d[str(key)] = value
    aviation_dict_list.append(d)
aviation_dict_list
```

```python
lax_dict = []
for d in aviation_dict_list:
    if d['Accident Number'] == 'LAX94LA336':
        lax_dict.append(d)
lax_dict
```

```python
count = 0
location = {}
for x in range(0,len(aviation_dict_list)): 
    state = aviation_dict_list[x]['Location'][-2:]
    if state in location.keys():
        location[state] += 1
    else:
        location[state] = 1
```

```python
month = {}
day = {}
for x in range(0,len(aviation_dict_list)): 
    m = aviation_dict_list[x]['Event Date'][0:2]
    d = aviation_dict_list[x]['Event Date'][3:5]
    if m in month.keys():
        month[m] += 1
    else:
        month[m] = 1
    if d in day.keys():
        day[d] += 1
    else:
        day[d] = 1
```
